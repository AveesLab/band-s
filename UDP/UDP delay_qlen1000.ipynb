```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
df1 = pd.read_csv('udp_qlen1000_normal.csv')
df2 = pd.read_csv('udp_qlen1000_nonblock.csv')
df3 = pd.read_csv('no dos.csv')
```

```python
# Creating a combined plot
plt.figure(figsize=(10, 7.1))

# Find the index where x-axis value is 22 or the nearest below 22
index_at_22 = df1['N Socket'].searchsorted(22, side='right') - 1

# Plotting df1 - Avg Delay up to x=22
plt.plot(df1['N Socket'][:index_at_22+1], df1['Avg Delay'][:index_at_22+1], marker='o', markersize=13, linewidth=4, color='orangered', label='Normal')

# Plotting df1 - Avg Delay from x=22 onwards as dashed line
plt.plot(df1['N Socket'][index_at_22:], df1['Avg Delay'][index_at_22:], marker='o', markersize=13, linewidth=4, color='orangered')

# Plot the average delay for the nonblock dataset
plt.plot(df2['N Socket'], df2['Avg Delay'], marker='o', markersize=13, linewidth=4, color='royalblue', label='Secure (non block)')

# Plot the average delay for the no_dos dataset
plt.plot(df3['N Socket'], df3['Avg Delay'], marker='o', markersize=13, linewidth=4, color='darkolivegreen', label='Secure (block)')

# Extend x-axis limit to 30 and start from 0
plt.xlim(0, 30)

# Adjusting y-axis limits based on the max delay values of both datasets
y_max = max(max(df1['max']), max(df1['Avg Delay']))
plt.ylim(bottom=0, top=40)

# Adding a vertical line at x = 22
#plt.axvline(x=22, color='orangered', linestyle='-', linewidth=4, alpha=1.0, label='Packet drop threshold', zorder=-100)

# Labeling the position x=4 on the x-axis
#plt.text(22.2, -plt.gca().get_ylim()[1]*0.091, '22', color='orangered', horizontalalignment='center', fontsize=40)

# Setting fontsize of x-axis and y-axis tick labels
plt.xticks(fontsize=40)
plt.yticks([0, 10, 20, 30, 40], fontsize=40)  # Include the tick at 40

plt.xlabel('Number of sockets', fontsize=55)
plt.ylabel('Delay (ms)', fontsize=55)

# Adding legend
#plt.legend(fontsize=23)

plt.grid(True)

plt.tight_layout()

# save image
plt.savefig('UDP_Delay (qlen1000).svg', format='svg')
plt.show()
```
