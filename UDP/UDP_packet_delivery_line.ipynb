```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
plt.rcParams['font.family'] = 'Times New Roman'
```

```python
df1 = pd.read_csv('UDP_packet_loss/udp_qlen500_suc.csv')
df2 = pd.read_csv('UDP_packet_loss/udp_qlen1000_suc.csv')
df3 = pd.read_csv('UDP_packet_loss/tcp_loss - 복사본.csv')
```

```python
plt.figure(figsize=(10, 7.1))

# Set the width of a bar
bar_width = 0.4

# Set positions of the bars
positions_df1 = df1['N Socket'] - bar_width / 2
positions_df2 = df2['N Socket'] + bar_width / 2

# Plotting data
plt.bar(positions_df1, df1['Packet Loss'], width=bar_width, color='#FB978D', alpha=1, label='UDP (q = 500)',
        edgecolor='dimgrey', linewidth=1.5, zorder=10)
plt.bar(positions_df2, df2['Packet Loss'], width=bar_width, color='cornflowerblue', alpha=1, label='UDP (q = 1000)',
        edgecolor='dimgrey', linewidth=1.5, zorder=20)
plt.plot(df3['N Socket'], df3['Packet Loss'], linewidth=3, color='green', label='TCP', zorder=200)
plt.axhline(y=0, xmin=0, xmax=30, color='green', linestyle='-', linewidth=2.8, zorder=500) # Adjusted for the axes

# Set labels
plt.xlabel('Number of sockets', fontsize=55)
plt.ylabel('Delivery Ratio (%)', fontsize=50)

# Adding legend
#plt.legend(fontsize=20, loc='upper right')
plt.legend(loc='lower right', bbox_to_anchor=(1.0,1.0), fontsize=20, ncol=3)

# Setting fontsize of x-axis and y-axis tick labels
#plt.xticks(df1['N Socket'], fontsize=10)
plt.xticks(np.arange(15, 31, 3), fontsize=40)
plt.yticks(fontsize=40)

# Adding grid
plt.grid(True)

# Set x and y limits
plt.xlim(14, 30.5)
plt.ylim(-5, 100)

# Invert y-axis if needed
# plt.gca().invert_yaxis()

plt.tight_layout()

# save image
plt.savefig('UDP_delivery_ratio.pdf', format='pdf')
plt.show()
```
