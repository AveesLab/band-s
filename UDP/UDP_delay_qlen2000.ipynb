```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
plt.rcParams['font.family'] = 'Times New Roman'

df1 = pd.read_csv('udp_qlen500_nonblock.csv')
df2 = pd.read_csv('udp_qlen500_block.csv')
df3 = pd.read_csv('no dos_qlen500.csv')

# Creating a combined plot
plt.figure(figsize=(10, 6))

# Find the index where x-axis value is 22 or the nearest below 22
index_at_17 = df1['N Socket'].searchsorted(17, side='right') - 1

# Plotting df1 - Avg Delay up to x=17
plt.plot(df1['N Socket'][:index_at_17+1], df1['Avg Delay'][:index_at_17+1], marker='o', linewidth=2.5, color='red', label='Normal')

# Plotting df1 - Avg Delay from x=17 onwards as dashed line
plt.plot(df1['N Socket'][index_at_17:], df1['Avg Delay'][index_at_17:], marker='o', linestyle='--', linewidth=2.5, color='red')

# Plot the average delay for the nonblock dataset
plt.plot(df2['N Socket'], df2['Avg Delay'], marker='o', linewidth=2.5, color='blue', label='Secure (non block)')

# Plot the average delay for the no_dos dataset
plt.plot(df3['N Socket'], df3['Avg Delay'], marker='o', linewidth=2.5, color='green', label='Secure (block)')

# Extend x-axis limit to 30 and start from 0
plt.xlim(0, 30)

# Adding a vertical line at x = 17
plt.axvline(x=17, color='goldenrod', linestyle='-', linewidth=2.5, alpha=0.7, label='Packet loss threshold', zorder=-100)

# Labeling the position x=17 on the x-axis
plt.text(17, -plt.gca().get_ylim()[1]*0.198, '17', color='goldenrod', horizontalalignment='center', fontsize=40)

# Setting fontsize of x-axis and y-axis tick labels
plt.xticks(fontsize=40)
plt.yticks([0, 10, 20, 30, 40], fontsize=40)  # Include the tick at 40

plt.xlabel('Number of Sockets', fontsize=45)
plt.ylabel('Delay (ms)', fontsize=45)

# Adding legend
plt.legend(fontsize=23)

plt.grid(True)

plt.tight_layout()

# save image
plt.savefig('UDP Delay (qlen500).pdf', format='pdf')
plt.show()
```
